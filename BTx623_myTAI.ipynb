```r
# load expression data 
exp <- read.table("./myTAI_BTx623/BTx623_RPKM_spline1000.txt", sep="\t", header=T)
head(exp)
```

```r
#rename sorghum genes to trim ".v3.2"
# current SbGIDS
old.rownames <- rownames(exp)
head(old.rownames)

# chop off extra bit
simple.rownames <- substr(old.rownames,1,nchar(old.rownames)-5)
head(simple.rownames)
```

```r
#rename
rownames (exp) <- simple.rownames
head(rownames(exp))
```

**Problem** : Query IDs for phylostrata are peptides. 
Divergence strata are transcripts.  
But gene expression is calculated at the gene model level.  

*Try* : Matching expression data with the highest strata from associated transcripts or peptides.  
This assumes that the most conserved product is the true ortholog.  


This is a mess, some genes are GRMZMXGXXXXX_PXX generated from GRMZMXGXXXXX  
others are ACXXXXXX.X_FGPXXX generated from ACXXXXXX.X_FGXXX... and there are even more patterns!  
I will manually rename peptides and transcripts in Excel,  
Reload duplicated PS and DS entries into R, 
and take the highest strata to make an expression dataframe.  


```r
ps_order <- ps[ order(ps$qseqid, ps$ps) , ]              # sort by id and reverse of abs(value)
head(ps_order)
dim(ps_order)
ps_deDup <- ps_order[ !duplicated(ps_order$qseqid) , ]   # take the first row within each id
head(ps_deDup)
dim(ps_deDup)
```

```r
#match phylostrata with expression data
#header from example: 
#Phylostratum      GeneID    Zygote  Quadrant  Globular    Heart  Torpedo     Bent   Mature

ps_merge <- merge(x = ps_deDup[, c("qseqid", "ps")], y = exp , by.x = "qseqid", by.y = "row.names")
head(ps_merge)
ps_exp <- ps_merge[, c(2,1,3:ncol(ps_merge))]
head(ps_exp)
```

```r
#try myTAI format checker?
is.ExpressionSet(ps_exp) 
```

```r
library('local.extrema')
```

```r
tai_vector <- TAI(ps_exp)
length(tai_vector)
head(tai_vector)
```

```r
extreme <- extrema(tai_vector, 100)
plot_extrema(extreme)
print(extreme)
extreme$maxima
extreme$minima
```

```r
dim(ps_exp)
head(ps_exp)
```

```r
library(ggplot2)
# Plot the Transcriptome Age Index of BTx623 spline data 
# Test Statistic : Flat Line Test (default)
p <- PlotSignature( ExpressionSet = ps_exp,
               measure       = "TAI", 
               TestStatistic = "FlatLineTest",
               xlab          = "Expression Time Units", 
               ylab          = "TAI",
               lwd           = 1 ) +
             theme(axis.text.x=element_blank(),
                   panel.grid.major = element_blank(), 
                   panel.grid.minor = element_blank(),
                   axis.line = element_line(colour = "black")) + 
             ylim(c(5.35, 5.46))
p

ggsave(filename = "./myTAI_BTx623/BTx623_TAI_profile_ggplot.pdf",
              plot = p,
              height = 3,
              width = 4)
```

```r
# Plot the Transcriptome Age Index of BTx623 spline data 
# Test Statistic : Flat Line Test (default)
PlotPattern( ExpressionSet = ps_exp,
               TestStatistic = "FlatLineTest",
               xlab          = "ETU", 
               ylab          = "TAI",
               lwd           = 1 )
```

```r
# save TAI profile
pdf(file = "./myTAI_BTx623/BTx623_TAI_profile.pdf",
        he = 3,
        wi = 4)

PlotPattern( ExpressionSet = ps_exp,
               TestStatistic = "FlatLineTest",
               xlab          = "ETU", 
               ylab          = "TAI",
               lwd           = 1 )

dev.off()
```

```r
# plot absolute contribution to expression
p <- PlotMeans(ps_exp, 
          Groups = list(c(1:16)),
          legendName = "PS",
          adjust.range = TRUE,
          lwd = 1,
          )  +
             theme(axis.text.x=element_blank(),
                   panel.grid.major = element_blank(), 
                   panel.grid.minor = element_blank(),
                   axis.line = element_line(colour = "black"))  
p

ggsave(filename = "./myTAI_BTx623/BTx623_PS_absolute_ggplot.pdf",
              plot = p,
              height = 3,
              width = 4)
```

```r
# plot relative expression
p <- PlotRE(ps_exp,
          Groups = list(c(1:16)), 
          legendName = "PS",
          adjust.range = TRUE,
          lwd = 1 ) +
             theme(axis.text.x=element_blank(),
                   panel.grid.major = element_blank(), 
                   panel.grid.minor = element_blank(),
                   axis.line = element_line(colour = "black"))  
p

ggsave(filename = "./myTAI_BTx623/BTx623_PS_relative_ggplot.pdf",
              plot = p,
              height = 3,
              width = 4)
```

```r
# load divergence strata
ds <- read.table("./myTAI_BTx623/BTx623_DS_SItalica_orthologr_dupGIDs.txt", sep="\t", header=T)
head(ds)
```

```r
ds_order <- ds[ order(ds$query_id, ds$DS) , ]              # sort by id and reverse of abs(value)
head(ds_order)
dim(ds_order)
ds_deDup <- ds_order[ !duplicated(ds_order$query_id) , ]   # take the first row within each id
head(ds_deDup)
dim(ds_deDup)
```

```r
#match divergence strata with expression data
#header from example: 
#DivergenceStrata      GeneID    Zygote  Quadrant  Globular    Heart  Torpedo     Bent   Mature

ds_merge <- merge(x = ds_deDup[, c("query_id", "DS")], y = exp , by.x = "query_id", by.y = "row.names")
head(ds_merge)
ds_exp <- ds_merge[, c(2,1,3:ncol(ds_merge))]
head(ds_exp)
```

```r
library('local.extrema')
```

```r
tdi_vector <- TAI(ds_exp)
length(tdi_vector)
head(tdi_vector)
```

```r
extreme <- extrema(tdi_vector, 100)
plot_extrema(extreme)
print(extreme)
extreme$maxima
extreme$minima
```

```r
dim(ds_exp)
head(ds_exp)
```

```r
# Test Statistic : Reductive Hourglass Test

ReductiveHourglassTest(ExpressionSet = ds_exp,
                       permutations  = 10000,
                       modules       = list(early = 1:220, mid = 221:583, late = 584:1000),
                       lillie.test  = T, 
                       runs = 100)
```

```r
# Plot the Transcriptome Age Index of BTx623 spline data 
# Test Statistic : Flat Line Test (default)
p <- PlotSignature( ExpressionSet = ds_exp,
               measure       = "TDI", 
               TestStatistic = "FlatLineTest",
               xlab          = "Expression Time Units", 
               ylab          = "TDI",
               lwd           = 1 ) +
             theme(axis.text.x=element_blank(),
                   panel.grid.major = element_blank(), 
                   panel.grid.minor = element_blank(),
                   axis.line = element_line(colour = "black")) + 
             ylim(c(4.855, 4.902))

p

ggsave(filename = "./myTAI_BTx623/BTx623_TDI_profile_ggplot.pdf",
              height = 3,
              width = 4)
```

```r
# Plot the Transcriptome Divergence Index of a given DivergenceExpressionSet 
# Test Statistic : Flat Line Test (default)
PlotPattern( ExpressionSet = ds_exp,
               TestStatistic = "FlatLineTest",
               xlab          = "Ontogeny", 
               ylab          = "TDI",
               lwd = 1)
```

```r
# save TDI profile
pdf(file = "./myTAI_BTx623/BTx623_TDI_profile.pdf",
        he = 3,
        wi = 4)

PlotPattern( ExpressionSet = ds_exp,
               TestStatistic = "FlatLineTest",
               xlab          = "Ontogeny", 
               ylab          = "TDI",
               lwd = 1)

dev.off()
```
